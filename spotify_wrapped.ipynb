```python
def TopSongs_2023():
    sql = """
                 select
                        track_name,
                        artist_name,
                        in_playlists,
                        in_charts,
                        streams
                 from spotify_2023
                 where released_year = 2023
                 order by streams desc limit 10
          """
    cur.execute(sql)
    result = cur.fetchall()
    # return result
    for i in result:
        # print(i[0],i[1],i[2],i[3],i[4])
       # print(f"{i[0]:20}{i[1]:20}{i[2]:20}{i[3]:20}{i[4]:20}")
        print(i)

def TopArtists_2023():
    sql = """
                select
                    artist_name,
                    sum(streams) as streams,
                    sum(in_playlists) as playlist_count,
                    sum(in_charts) as charts_count
                from spotify_2023
                where released_year = 2023
                group by artist_name
                order by streams desc limit 10
          """
    cur.execute(sql)
    result = cur.fetchall()
    for i in result:
        print(i)
        
def most_songs_count_by_artist():
    sql = """
                SELECT 
                  artist_name,
                  COUNT(*) as song_count
                FROM spotify_2023
                WHERE released_year = 2023
                GROUP BY artist_name
                ORDER BY song_count DESC limit 10
          """
    cur.execute(sql)
    result = cur.fetchall()
    for i in result:
        print(i)

def average_number_of_streams():
    sql = """
                SELECT
                      released_month,
                      monthname(concat("2023-",released_month,"-01")) as month,
                      ROUND(AVG(streams),0) as avg_streams
                FROM spotify_2023
                WHERE released_year = 2023
                GROUP BY released_month
                ORDER BY released_month  
          """
    cur.execute(sql)
    result = cur.fetchall()
    for i in result:
        print(i)
```
